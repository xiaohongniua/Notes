```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
X=np.array([np.linspace(-5,5,20) for i in range(10)])
```

```python
X
```

```python
Y=X**2+2
```

```python
Y
```

# Round 1

```python
for i in range(10):
    plt.cla()
    plt.plot(X[i],Y[i],c='#%sAD631'%i)
    plt.show()
```

you can get 10 pictures

# round2

```python
for i in range(10):
    plt.cla()
    plt.plot(X[i],Y[i],c='#%sAD631'%i)
plt.show()
```

There is only one picture in the result.

# round3

```python
for i in range(10):
    plt.cla()
    plt.plot(X[i],Y[i],c='#%sAD631'%i)
    plt.draw()
```

# round4

```python
for i in range(10):
    plt.cla()
    plt.plot(X[i],Y[i],c='#%sAD631'%i)
    plt.draw()
plt.show()
```

```python
for i in range(10):
    plt.cla()
    plt.plot(X[i],Y[i],c='#%sAD631'%i)
    plt.draw()
    plt.pause(0.05)
plt.show()
```

In block mode you can use plt.ion() and plt.ioff() to show dynamic pictures()
